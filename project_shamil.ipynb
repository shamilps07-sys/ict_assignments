```python
#9 Text-Based Decision Game
Create a text-based decision-making game.
The game should present players with scenarios where choices affect the storyline and
score. Multiple story paths and endings must exist.
The system should track decisions, compute scores, and allow replayability. String
processing and conditional branching are core requirements.
```

```python
 score = 0

print("WELCOME TO THE ADVENTURE GAME")
print("You are at the entrance of a forest.")

print("1. Enter the forest")
print("2. Go back home")

choice1 = input("Enter your choice (1 or 2): ")

if choice1 == "1":
    score = score + 10
    print("You entered the forest.")

    print("You see a river.")
    print("1. Drink water")
    print("2. Walk away")

    choice2 = input("Enter your choice (1 or 2): ")

    if choice2 == "1":
        score = score + 10
        print("You feel refreshed.")
        print("GOOD ENDING: You survived the forest.")

    else:
        score = score + 5
        print("You got lost but found your way out.")
        print("NORMAL ENDING.")

else:
    print("You went back home safely.")
    print("SAFE ENDING.")

print("Your final score is:", score)
```

```python
4 '''''Mini Banking System with Transaction History
Design a Mini Banking System that simulates basic banking operations.
The system should support account creation, deposits, withdrawals, and fund transfers
between accounts. Each transaction must be recorded and retrievable as a mini statement.
Business rules include minimum balance enforcement, daily withdrawal limits, and
rejection of invalid transactions. Data persistence during runtime and menu-based
navigation are mandatory
```

```python
# Initial data
balance = 1000
transactions = []
daily_limit = 5000

while True:
    print("\n--- MINI BANK MENU ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Transaction History")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    # Deposit
    if choice == "1":
        amount = int(input("Enter deposit amount: "))
        if amount > 0:
            balance = balance + amount
            transactions.append("Deposited: " + str(amount))
            print("Deposit successful")
        else:
            print("Invalid amount")

    # Withdraw
    elif choice == "2":
        amount = int(input("Enter withdraw amount: "))
        if amount > 0 and amount <= balance and amount <= daily_limit:
            balance = balance - amount
            transactions.append("Withdrawn: " + str(amount))
            print("Withdrawal successful")
        else:
            print("Invalid withdrawal")

    # Check balance
    elif choice == "3":
        print("Current Balance:", balance)

    # Transaction history
    elif choice == "4":
        print("\nTransaction History:")
        if len(transactions) == 0:
            print("No transactions yet")
        else:
            for t in transactions:
                print(t)

    # Exit
    elif choice == "5":
        print("Thank you for using the bank")
        break

    else:
        print("Invalid choice")
```
